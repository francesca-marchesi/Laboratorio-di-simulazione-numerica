In questa esercitazione è stato scritto un algoritmo genetico (GA) ed è stato utilizzato per risolvere un noto problema di minimizzazione chiamato Traveling Salesman Problem (TSP).
L'algoritmo scritto trova la sequenza con cui le città devono essere visitate in modo che la distanza totale percorsa sia minima. l'argoritmo è stato implementato come indicato nel jupiter notebook: i possibili percorsi sono rappresentati da vettori unidimensionali di interi e sono state costruite le funzioni di **Check** (verifica che i singoli elementi della popolazione soddisfino i vincoli del problema), l'**operatore di selezione** (in base alla funzione di fitness, implementata anch'essa, determina la probabilità che il singolo elemento della popolazione venga scelto) e le funzioni di **mutazione** e **crossover**.

Per quanto riguarda la funzione **Fitness** essa calcola la lunghezza totale del percorso scelto mediante la formula

$$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$

La funzione **Check** verifica che ogni singola città venga visitata una sola volta, che il numero che le corrisponde compaia una sola volta nella sequenza, 

L'operatore di selezione (**Selection**) calcola la probabilità che ogni sequenza ha di essere selezionata all'interno del campione basandosi sulla fintess che le corrisponde: sequenze con Fitness minore. La formula è la seguente:

$$
P_i=\frac{1}{L^{(1)}_i}\frac{1}{\sum_i L^{(1)}_i}
$$

gli operatori di mutazione e crossover sono quelli indicati nel Jupiter Notebook.

Per la distribuzione delle città sono state implementate due funzioni (**Square** e **Circle**) seguendo le indicazioni del Jupiter Notebook.
